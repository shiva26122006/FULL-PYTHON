## SET

```python
#  Create a set of 5 numbers and print it
set1 = {10, 20, 30, 40, 50}
print(set1)
```

```python
#  Create a set with duplicate values and print the result
set2 = {1, 2, 2, 3, 4, 4, 5}
print(set2)
```

```python
#  Create an empty set
empty_set = set()
print(empty_set)
```

```python
#  Add one element to a set using add()
set1.add(60)
print(set1)
```

```python
#  Add multiple elements to a set using update()
set1.update([70, 80, 90])
print(set1)
```

```python
#  Remove an element from a set using remove()
set1.remove(20)
print(set1)
```

```python
#  Remove an element from a set using discard()
set1.discard(30)
print(set1)
```

```python
#  Remove a random element from a set using pop()
set1.pop()
print(set1)
```

```python
#  Find the length of a set
print(len(set1))
```

```python
#  Perform union of two sets
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print(A.union(B))  #OR #print(A|B)
```

```python
#  Perform intersection of two sets
print(A.intersection(B))   #OR #print(A&B)
```

```python
#  Find the difference between two sets
print(A.difference(B))     #OR #print(A-B)
```

```python
#  Find the symmetric difference between two sets
print(A.symmetric_difference(B))     #OR #print(A^B)
```

## DICTIONARY

```python
#  Create a dictionary with 3 key-value pairs
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student)
```

```python
#  Access two values using keys
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student["name"])
print(student["age"])
```

```python
#  Add a new key-value pair
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
student["city"] = "goa"
print(student)
```

```python
# Update the value of an existing key
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
student["age"] = 21
print(student)
```

```python
#  Remove a key using pop()
student = {
    "name": "Shiva",
    "age": 19,
    "city": "goa"
}
student.pop("city")
print(student)
```

```python
#  Delete a key using del keyword
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
del student["course"]
print(student)
```

```python
# Find the length of a dictionary
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(len(student))
```

```python
#  Check whether a key exists in a dictionary
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print("name" in student)
```

```python
#  Use get() method to access a value
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student.get("name"))
```

```python
# Use keys() method
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student.keys())
```

```python
#  Use values() method
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student.values())
```

```python
#  Use items() method
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student.items())
```
