```python
#  Perform union of two sets
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print(A.union(B))  #OR #print(A|B)
```

```python
#  Perform intersection of two sets
print(A.intersection(B))   #OR #print(A&B)
```

```python
#  Find the difference between two sets
print(A.difference(B))     #OR #print(A-B)
```

```python
#  Find the symmetric difference between two sets
print(A.symmetric_difference(B))     #OR #print(A^B)
```

## DICTIONARY

```python
#  Create a dictionary with 3 key-value pairs
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student)
```

```python
#  Access two values using keys
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student["name"])
print(student["age"])
```

```python
#  Add a new key-value pair
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
student["city"] = "goa"
print(student)
```

```python
# Update the value of an existing key
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
student["age"] = 21
print(student)
```

```python
#  Remove a key using pop()
student = {
    "name": "Shiva",
    "age": 19,
    "city": "goa"
}
student.pop("city")
print(student)
```

```python
#  Delete a key using del keyword
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
del student["course"]
print(student)
```

```python
# Find the length of a dictionary
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(len(student))
```

```python
#  Check whether a key exists in a dictionary
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print("name" in student)
```

```python
#  Use get() method to access a value
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student.get("name"))
```

```python
# Use keys() method
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student.keys())
```

```python
#  Use values() method
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student.values())
```

```python
#  Use items() method
student = {
    "name": "Shiva",
    "age": 19,
    "course": "BTech"
}
print(student.items())
```

```python
# to print name and engter five subject marks and print the total , percentage and grade , reult.
name = input("Enter Name: ")

m1 = int(input("Enter the marks of first subject"))
m2 = int(input("Enter the marks of second subject"))
m3 = int(input("Enter the marks of third subject"))
m4 = int(input("Enter the marks of fourth subject"))
m5 = int(input("Enter the marks of fifth subject"))

total = m1 + m2 + m3 + m4 + m5    # to calculate the sum of all five subjects marks.
percentage = total / 5           #  to calculate the percentage.

if percentage >= 90:
    grade = "A"
elif percentage >= 75:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 40:
    grade = "D"
else:
    grade = "F"

if percentage >= 40:
    result = "Pass"
else:
    result = "Fail"

print("Name:", name)
print("Total:", total,"/500")
print("Percentage:", percentage)
print("Grade:", grade)
print("Result:", result)

```
